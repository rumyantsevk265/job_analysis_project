## Определим доли грейдов Junior, Junior+, Middle, Senior среди вакансий Аналитик данных и Специалист по Data Science.


Убедимся какие грейды указывали в вакансиях.

```python
df_da['experience'].unique()
```

```python
df_ds['experience'].unique()
```

Как мы можем видеть, среди уникальных значений указаны четыре нужные нам разделения по опыту кандидата.

Напишем функцию, которая классифицирует уровень квалификации кандидата на основе его опыта работы. Функция принимает строку с описанием опыта работы и возвращает соответствующий грейд:

    'Senior' для опыта работы более 6 лет,
    'Middle' для опыта работы от 3 до 6 лет,
    'Junior+' для опыта работы от 1 до 3 лет,
    'Junior' для кандидатов без опыта работы.
    Если опыт работы не указан или не соответствует ни одной из категорий, возвращается 'Not Specified'.

```python
# функция для класификации уровня квалификации кандидата
def determine_grade(experience):
    if experience == 'Senior (6+ years)':
        return 'Senior'
    elif experience == 'Middle (3-6 years)':
        return 'Middle'
    elif experience == 'Junior+ (1-3 years)':
        return 'Junior+'
    elif experience == 'Junior (no experince)':
        return 'Junior'
    else:
        return 'Not Specified'
```

Посмотрим грейд для датафрейма с вакансиями аналитика данных:

```python
# Применение функции к столбцу 'experience' датафрейма df_da
df_da['grade_da'] = df_da['experience'].apply(determine_grade)

# Подсчет количества каждого грейда
grade_counts_da = df_da['grade_da'].value_counts()

# Расчет доли каждого грейда
grade_shares_da = grade_counts_da / grade_counts_da.sum()

grade_shares_da
```

```python
# Данные для визуализации
grades = ['Junior+', 'Middle', 'Junior', 'Senior']
shares = grade_counts_da.values  # Используйте реальные данные здесь

# Создание круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(shares, autopct='%1.1f%%', startangle=140)

# Добавление легенды
plt.legend(grades, title="Грейды", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Распределение долей грейдов кандидатов на вакансии DA')
plt.show()
```

Посмотрим грейд для датафрейма с вакансиями специалиста по DS

```python
# Применение функции к столбцу 'experience' датафрейма df_ds
df_ds['grade_ds'] = df_ds['experience'].apply(determine_grade)

# Подсчет количества каждого грейда
grade_counts_ds = df_ds['grade_ds'].value_counts()

# Расчет доли каждого грейда
grade_shares_ds = grade_counts_ds / grade_counts_ds.sum()

grade_shares_ds
```

```python
# Данные для визуализации
grades = ['Middle','Junior+', 'Junior', 'Senior']
shares = grade_counts_ds.values  # Используйте реальные данные здесь

# Создание круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(shares, autopct='%1.1f%%', startangle=140)

# Добавление легенды
plt.legend(grades, title="Грейды", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Распределение долей грейдов кандидатов на вакансии специалист DS')
plt.show()
```

**Вывод:**

Вывод по долям грейдов среди вакансий показывает, что на рынке труда для аналитиков данных и специалистов по Data Science преобладают вакансии для специалистов со средним уровнем опыта (Middle) и начинающих специалистов с небольшим опытом (Junior+).

Для **Аналитиков данных**:
- Большинство вакансий (57%) предназначены для специалистов с небольшим опытом работы (Junior+).
- Следующая по величине категория (39%) - это вакансии для специалистов со средним уровнем опыта (Middle).
- Доля вакансий для начинающих специалистов без опыта (Junior) и для высококвалифицированных специалистов (Senior) составляет всего 3% и 1% соответственно.

Для **Специалистов по Data Science**:
- Наибольшая доля вакансий также приходится на специалистов со средним уровнем опыта (Middle) - 49%.
- Вакансии для специалистов с небольшим опытом работы (Junior+) составляют 44%.
- Начинающие специалисты без опыта (Junior) и высококвалифицированные специалисты (Senior) занимают 4% и 3% рынка соответственно.

Эти данные указывают на то, что рынок труда в области анализа данных и Data Science ориентирован на специалистов, которые уже имеют определенный уровень опыта, но еще не достигли статуса экспертов. Это может быть связано с потребностью в специалистах, способных быстро адаптироваться и развиваться в быстро меняющейся сфере, а также с желанием работодателей инвестировать в развитие персонала. ### Теперь посмотрим на наиболее желаемые навыки в вакансиях в разрезе грейдов: Junior, Junior+, Middle, Senior.

```python
# Функция для определения наиболее важных навыков по грейдам для двух датафреймов df_da, df_ds
def find_top_skills_by_grade(df, column, grade_column, grade, top_n=5):
    # Фильтруем датафрейм по грейду
    df_grade = df[df[grade_column] == grade]
    # Собираем все навыки в один список, игнорируя пустые строки и строки только с запятыми
    all_skills = []
    for skills in df_grade[column]:
        if pd.notnull(skills) and skills.strip(','):
            all_skills.extend([skill.strip() for skill in skills.split(',') if skill.strip()])
    # Преобразуем в Series и считаем количество упоминаний каждого навыка
    skills_series = pd.Series(all_skills).value_counts()
    # Возвращаем топ-N наиболее часто упоминаемых навыков
    return skills_series.head(top_n).index.tolist()

# Предполагаем, что функция determine_grade уже определена
# Применяем функцию determine_grade к столбцу 'experience' для обоих датафреймов
df_da['grade_da'] = df_da['experience'].apply(determine_grade)
df_ds['grade_ds'] = df_ds['experience'].apply(determine_grade)

grades = ['Junior', 'Junior+', 'Middle', 'Senior']

# Выводим топ hard-skills и soft-skills для каждого грейда для df_da
for grade in grades:
    print(f'Топ hard-skills для {grade} в df_da:')
    print(find_top_skills_by_grade(df_da, 'hard_skills_from_description', 'grade_da', grade))
    print(f'Топ soft-skills для {grade} в df_da:')
    print(find_top_skills_by_grade(df_da, 'soft_skills_from_description', 'grade_da', grade))
    print('\n' + '-'*50 + '\n')
```

```python
# Выводим топ hard-skills и soft-skills для каждого грейда для df_ds
for grade in grades:
    print(f'Топ hard-skills для {grade} в df_ds:')
    print(find_top_skills_by_grade(df_ds, 'hard_skills_from_description', 'grade_ds', grade))
    print(f'Топ soft-skills для {grade} в df_ds:')
    print(find_top_skills_by_grade(df_ds, 'soft_skills_from_description', 'grade_ds', grade))
    print('\n' + '-'*50 + '\n')
```

## Определение типичного места работы для Аналитика данных и специалист по Data Science по следующим параметрам: ТОП-работодателей, зарплата, тип занятости, график работы. Отдельно рассмотрим в разрезе для грейдов Junior, Junior+, Middle, Senior.


```python
# Функция для анализа данных по параметрам и грейдам
def analyze_job_market(df, grade):
    # Фильтрация данных по грейду
    df_filtered = df[df['experience'].str.contains(grade)]

    # ТОП-работодателей
    top_employers = df_filtered['employer'].value_counts().head(5)

    # Распространенный тип занятости
    common_employment_type = df_filtered['employment'].mode()[0]

    # Распространенный график работы
    common_schedule = df_filtered['schedule'].mode()[0]

    return top_employers, common_employment_type, common_schedule

grades = ['Junior', 'Junior+', 'Middle', 'Senior']

# Анализ для каждого грейда
for grade in grades:
    print(f"Анализ для грейда {grade}:\n")

    # Анализ для Аналитиков данных
    top_employers_da, common_employment_type_da, common_schedule_da = analyze_job_market(df_da, grade)
    print("Аналитики данных:")
    print(f"ТОП-5 работодателей:\n{top_employers_da}")
    print(f"Типичный тип занятости: {common_employment_type_da}")
    print(f"Типичный график работы: {common_schedule_da}\n")

    # Анализ для Специалистов по Data Science
    top_employers_ds, common_employment_type_ds, common_schedule_ds = analyze_job_market(df_ds, grade)
    print("Специалисты по Data Science:")
    print(f"ТОП-5 работодателей:\n{top_employers_ds}")
    print(f"Типичный тип занятости: {common_employment_type_ds}")
    print(f"Типичный график работы: {common_schedule_ds}\n")
```

```python

# Функция для отображения результатов анализа рынка труда
def visualize_job_market_analysis(df_da, df_ds, grades):
    for grade in grades:
        # Получение данных для анализа
        top_employers_da, employment_type_da, schedule_da = analyze_job_market(df_da, grade)
        top_employers_ds, employment_type_ds, schedule_ds = analyze_job_market(df_ds, grade)

        # Создание фигуры и осей для графиков
        fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

        # Визуализация ТОП-5 работодателей для Аналитиков данных
        axes[0].bar(top_employers_da.index, top_employers_da.values, color='lightblue')
        axes[0].set_title(f'ТОП-5 работодателей для грейда {grade} (Аналитики данных)')
        axes[0].set_ylabel('Количество вакансий')
        axes[0].tick_params(axis='x', rotation=45)

        # Визуализация ТОП-5 работодателей для Специалистов по Data Science
        axes[1].bar(top_employers_ds.index, top_employers_ds.values, color='orange')
        axes[1].set_title(f'ТОП-5 работодателей для грейда {grade} (Data Science)')
        axes[1].set_ylabel('Количество вакансий')
        axes[1].tick_params(axis='x', rotation=45)

        # Добавление текста с типичным типом занятости и графиком работы
        fig.text(0.5, 0.04, f'Типичный тип занятости: {employment_type_da} (Аналитики данных), {employment_type_ds} (Data Science)', ha='center')
        fig.text(0.5, 0.00, f'Типичный график работы: {schedule_da} (Аналитики данных), {schedule_ds} (Data Science)', ha='center')

        # Отображение графика
        plt.tight_layout(rect=[0, 0.1, 1, 1])
        plt.show()

# Пример использования функции
visualize_job_market_analysis(df_da, df_ds, grades)
```
